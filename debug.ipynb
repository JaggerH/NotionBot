```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd

def get_table_from_url(stock_code):
    # 设置Chrome驱动
    service = Service('/Users/jhuang/Downloads/chromedriver_mac64')  # 请将'path_to_chromedriver'替换为你本地的chromedriver路径
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    driver = webdriver.Chrome(service=service, options=options)
    
    # 构造URL
    url = f"https://emweb.securities.eastmoney.com/pc_hsf10/pages/index.html?type=web&code={stock_code}&color=b#/jyfx"
    
    # 访问URL
    driver.get(url)
    
    # 等待表格加载
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.TAG_NAME, 'table'))
        )
    except Exception as e:
        print(f"Error: {e}")
        driver.quit()
        return None
    
    # 获取页面内容
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    # 查找第一个表格
    table = soup.find_all('table')[0]
    
    # 提取表格内容
    headers = [header.text for header in table.find_all('th')]
    rows = []
    for row in table.find_all('tr'):
        rows.append([cell.text for cell in row.find_all('td')])
    
    # 创建DataFrame
    df = pd.DataFrame(rows, columns=headers)
    
    # 关闭浏览器
    driver.quit()
    
    return df

# 示例股票代码
stock_code = 'SH600104'
table_df = get_table_from_url(stock_code)

# 显示表格内容
print(table_df)
```
